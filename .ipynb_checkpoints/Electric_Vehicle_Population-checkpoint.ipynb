```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ev_population = pd.read_csv("../Resources/Electric_Vehicle_Population_Data.csv")

ev_population
```

```python
clean_df = ev_population.drop(columns = ['VIN (1-10)','Base MSRP','Postal Code','Legislative District', 'DOL Vehicle ID','2020 Census Tract','County', 'Electric Utility'])
display(clean_df.head())
display(clean_df.tail())
```

```python
clean_df = clean_df.sort_values(by=['Model Year', 'City'], ascending=True)
clean_df = clean_df.rename(columns={'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV'})
display(clean_df.head())
```

```python
# Filter the original DataFrame
filtered_df = clean_df.loc[(clean_df['Model Year'] >= 2014) & (clean_df['Model Year'] <= 2019)]

# Group the filtered DataFrame
city_df = filtered_df.groupby('Model Year')['Electric Range'].mean().reset_index()
data = pd.DataFrame({'Model Year': [2014, 2015, 2016, 2017, 2018, 2019],
                     'Electric Range': [80.238435, 97.784441, 101.284676, 114.242376, 156.440207, 176.720564]})

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(data['Model Year'], data['Electric Range'])
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Electric Range', fontsize=12)
plt.xticks(data['Model Year'])
plt.tight_layout()
plt.show()

city_df
```
